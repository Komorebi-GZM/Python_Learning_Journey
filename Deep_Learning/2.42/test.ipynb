```python
# ==================== d2l 深度学习环境测试 ====================
import torch
import torch.nn as nn

# 测试1：PyTorch 基础
print("=== 测试1：PyTorch 版本 & CUDA 可用性 ===")
print(f"PyTorch 版本：{torch.__version__}")
print(f"CUDA 是否可用：{torch.cuda.is_available()}")
if torch.cuda.is_available():
    print(f"CUDA 设备数：{torch.cuda.device_count()}")
    print(f"当前 CUDA 设备：{torch.cuda.get_device_name(0)}")

# 测试2：张量运算
print("\n=== 测试2：张量基础运算 ===")
x = torch.randn(3, 3)
y = torch.ones(3, 3)
print("张量 x：")
print(x)
print("\n张量 x + y：")
print(x + y)
print("\n张量 x 矩阵乘法：")
print(x @ y)

# 测试3：简单神经网络
print("\n=== 测试3：简单神经网络定义 ===")
class SimpleNet(nn.Module):
    def __init__(self):
        super().__init__()
        self.linear = nn.Linear(10, 1)
    
    def forward(self, x):
        return self.linear(x)

net = SimpleNet()
print("神经网络结构：")
print(net)

# 测试4：验证环境路径
print("\n=== 测试4：当前环境路径 ===")
import sys
print(f"Python 解释器路径：{sys.executable}")
print("\n✅ d2l 深度学习环境测试通过！")
```
