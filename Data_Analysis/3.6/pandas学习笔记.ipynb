### 2.1 重新索引

```python
# reindex创建新对象，数据按新索引排列
obj = pd.Series([4.5, 7.2, -5.3, 3.6], index=['d', 'b', 'a', 'c'])
obj
```

```python
obj2 = obj.reindex(['a', 'b', 'c', 'd', 'e'])
obj2
```

```python
# method参数处理缺失值，ffill向前填充
obj3 = pd.Series(['blue', 'purple', 'yellow'], index=[0, 2, 4])
obj3
```

```python
obj3.reindex(range(6), method='ffill')
```

```python
# DataFrame的reindex
frame = pd.DataFrame(np.arange(9).reshape((3, 3)),
                     index=['a', 'c', 'd'],
                     columns=['Ohio', 'Texas', 'California'])
frame
```

## 补充：实际数据处理示例

```python
# 模拟股票数据，计算相关系数和协方差
np.random.seed(42)
dates = pd.date_range('2023-01-01', periods=100)

# 模拟3只股票的收益率
returns = pd.DataFrame({
    'AAPL': np.random.normal(0.001, 0.02, 100),
    'GOOG': np.random.normal(0.0015, 0.025, 100),
    'MSFT': np.random.normal(0.0008, 0.018, 100)
}, index=dates)

returns.head()
```

```python
# 计算累计收益率
cumulative = (1 + returns).cumprod()
cumulative.head()
```

```python
# 相关系数矩阵
returns.corr()
```

```python
# 协方差矩阵
returns.cov()
```

```python
# 计算移动平均（20日）
returns_ma = returns.rolling(window=20).mean()
returns_ma.tail()
```

```python
# 统计描述
returns.describe()
```
