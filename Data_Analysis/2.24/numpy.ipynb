## 第1章：Pandas基础

### 1.1 导入Pandas和NumPy

```python
# 导入pandas并使用pd作为别名（约定俗成）
import pandas as pd
# 同时导入numpy，因为pandas底层依赖numpy
import numpy as np
```

### 1.2 Pandas数据结构 - Series

```python
# Series是一维数组对象，包含数据和索引
# 从列表创建Series，默认索引是0,1,2,...
obj = pd.Series([4, 7, -5, 3])
obj
```

```python
# 获取Series的值（返回numpy数组）
obj.values
```

```python
# 获取Series的索引对象
obj.index
```

```python
# 创建带有自定义索引的Series
obj2 = pd.Series([4, 7, -5, 3], index=['d', 'b', 'a', 'c'])
obj2
```

```python
# 通过索引标签选择数据
obj2['a']
```

```python
# 通过索引标签赋值
obj2['d'] = 6
obj2
```

```python
# 同时选择多个索引
obj2[['c', 'a', 'd']]
```

```python
# 使用布尔条件过滤
obj2[obj2 > 0]
```

```python
# Series与标量运算
obj2 * 2
```

```python
# 应用NumPy函数
np.exp(obj2)
```

```python
# 可以将Series看作固定长度的有序字典
'b' in obj2
```

```python
'e' in obj2
```

```python
# 从Python字典创建Series
sdata = {'Ohio': 35000, 'Texas': 71000, 'Oregon': 16000, 'Utah': 5000}
obj3 = pd.Series(sdata)
obj3
```

```python
# 传入索引列表，可以指定顺序，缺失值会用NaN填充
states = ['California', 'Ohio', 'Oregon', 'Texas']
obj4 = pd.Series(sdata, index=states)
obj4
```

```python
# 检查缺失值：isnull和notnull
pd.isnull(obj4)
```

```python
# Series的实例方法
obj4.isnull()
```

```python
# Series自动对齐索引进行运算
obj3
```

```python
obj4
```

```python
# 相加时，索引对齐，不匹配的为NaN
obj3 + obj4
```

```python
# Series及其索引都有name属性
obj4.name = 'population'
obj4.index.name = 'state'
obj4
```

```python
# 索引可以通过赋值就地修改
obj.index = ['Bob', 'Steve', 'Jeff', 'Ryan']
obj
```
