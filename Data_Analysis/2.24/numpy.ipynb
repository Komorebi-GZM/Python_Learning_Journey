## 第3章：布尔型索引

```python
# 创建名字数组和对应的数据
names = np.array(['Bob', 'Joe', 'Will', 'Bob', 'Will', 'Joe', 'Joe'])
data = np.random.randn(7, 4)
names
```

```python
data
```

```python
# 布尔比较：返回布尔数组
names == 'Bob'
```

```python
# 使用布尔数组索引：只选择对应True的行
# 这里选择名字是Bob的行
data[names == 'Bob']
```

```python
# 结合布尔索引和切片
# 选择Bob的行，从第3列开始
data[names == 'Bob', 2:]
```

```python
# 选择Bob的行，第4列
data[names == 'Bob', 3]
```

```python
# 使用!=或者~进行否定
names != 'Bob'
```

```python
data[~(names == 'Bob')]
```

```python
# 也可以先保存条件
cond = names == 'Bob'
data[~cond]
```

```python
# 使用|（或）和&（且）组合多个条件
# 注意：不能用and/or，必须用&/|
mask = (names == 'Bob') | (names == 'Will')
mask
```

```python
data[mask]
```

```python
# 使用布尔索引设置值
# 将所有负数设为0
data[data < 0] = 0
data
```

```python
# 将不是Joe的行全部设为7
data[names != 'Joe'] = 7
data
```

### 7.1 将条件逻辑表述为数组运算

```python
xarr = np.array([1.1, 1.2, 1.3, 1.4, 1.5])
yarr = np.array([2.1, 2.2, 2.3, 2.4, 2.5])
cond = np.array([True, False, True, True, False])
```

```python
# Python列表推导式实现三元运算符
# 缺点：对大数组慢，且不能用于多维数组
result = [(x if c else y)
          for x, y, c in zip(xarr, yarr, cond)]
result
```

```python
# 使用np.where：向量化的三元运算符
# np.where(condition, x, y)：condition为True选x，否则选y
result = np.where(cond, xarr, yarr)
result
```

```python
arr = np.random.randn(4, 4)
arr
```

```python
arr > 0
```

## 第11章：示例：随机漫步

```python
# Python纯循环实现随机漫步
import random
position = 0
walk = [position]
steps = 1000
for i in range(steps):
    step = 1 if random.randint(0, 1) else -1
    position += step
    walk.append(position)
```

```python
# 可视化前100步
plt.plot(walk[:100])
```

```python
# NumPy向量化实现（更高效）
nsteps = 1000
# 生成0或1
draws = np.random.randint(0, 2, size=nsteps)
# 转换为+1或-1
steps = np.where(draws > 0, 1, -1)
# 累加得到漫步轨迹
walk = steps.cumsum()
```

```python
# 统计信息：最小值
walk.min()
```

```python
# 最大值
walk.max()
```

```python
# 首次到达距离原点>=10的步数
# argmax返回第一个True的索引
(np.abs(walk) >= 10).argmax()
```

```python
# 一次模拟多次随机漫步（5000次）
nwalks = 5000
nsteps = 1000
draws = np.random.randint(0, 2, size=(nwalks, nsteps))
steps = np.where(draws > 0, 1, -1)
walks = steps.cumsum(1)
walks
```

```python
# 所有漫步的最大值
walks.max()
```

```python
# 所有漫步的最小值
walks.min()
```

```python
# 计算哪些漫步到达过距离原点>=30
hits30 = (np.abs(walks) >= 30).any(1)
hits30
```

```python
# 到达过30的漫步数量
hits30.sum()
```

```python
# 只看那些到达过30的漫步
# 计算它们首次到达30的平均步数
crossing_times = (np.abs(walks[hits30]) >= 30).argmax(1)
crossing_times.mean()
```
